# Resampling

```python
dfresampling = df.resample('A') # resampling annually
```

```python
dfresampling.mean()
```

```python
df.resample('M') # resamling monthly
```

```python
df.resample('M').mean()
```

```python
df.resample('Q').sum() # resampling quarterly
```

```python
df.resample('Q').sum().plot()
```

```python
df[df.isnull().all(axis=1)] # finding all null rows
```

```python
nullrows = df[df.isnull().all(axis=1)].index
```

```python
nullrows
```

```python
df.drop(index= nullrows,inplace = True)
```

```python
df.resample('W').mean()
```

```python
df.resample('Q').sum().plot(figsize = (15,12))
```

INSIGHTS:
From the above graph, we can see that rest4's customers are decreasing over the time which can be a red flag to the restaurant's business, however rest3 manages to keep the no. of customers constant till the first quarter of 2017.There is an increase in no. of customers for rest1 which is good sign. of improvement till Q1 of 2017.
